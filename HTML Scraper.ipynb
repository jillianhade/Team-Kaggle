```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
```

```python
biology = pd.read_csv('rawdata/biology.csv')
```

```python
cooking = pd.read_csv('rawdata/cooking.csv')
```

```python
crypto = pd.read_csv('rawdata/crypto.csv')
```

```python
diy = pd.read_csv('rawdata/diy.csv')
```

```python
robotics = pd.read_csv('rawdata/robotics.csv')
```

```python
travel = pd.read_csv('rawdata/travel.csv')
```

```python
biology = biology[['title', 'content', 'tags']].as_matrix()
cooking = cooking[['title', 'content', 'tags']].as_matrix()
crypto = crypto[['title', 'content', 'tags']].as_matrix()
diy = diy[['title', 'content', 'tags']].as_matrix()
robotics = robotics[['title', 'content', 'tags']].as_matrix()
travel = travel[['title', 'content', 'tags']].as_matrix()

def parse_html(data_np):    
    for i in range(data_np.shape[0]):
        soup = BeautifulSoup(data_np[i,1], 'html.parser')
        soup = soup.get_text()
        soup = BeautifulSoup(soup, 'html.parser')
        soup = soup.decode('utf8')
        data_np[i,1] = soup.replace('\n', ' ')

parse_html(biology)
parse_html(cooking)
parse_html(crypto)
parse_html(diy)
parse_html(robotics)
parse_html(travel)
```

```python
biology = pd.DataFrame(data=biology, index=None, columns=['title', 'text', 'tags'])
cooking = pd.DataFrame(data=cooking, index=None, columns=['title', 'text', 'tags'])
crypto = pd.DataFrame(data=crypto, index=None, columns=['title', 'text', 'tags'])
diy = pd.DataFrame(data=diy, index=None, columns=['title', 'text', 'tags'])
robotics = pd.DataFrame(data=robotics, index=None, columns=['title', 'text', 'tags'])
travel = pd.DataFrame(data=travel, index=None, columns=['title', 'text', 'tags'])
```
